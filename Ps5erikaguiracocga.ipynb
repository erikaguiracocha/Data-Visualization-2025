**Dataset Overview and Research Focus**

This study examines data from 21 counties in New Jersey between 2020 and 2022 to explore how local crime rates influence police budget allocation. By analyzing reported crimes, budgetary figures, and socioeconomic indicators—such as population and unemployment rates—this research aims to uncover patterns in how counties prioritize public safety spending in response to varying crime levels.

**Research Story:** Why This Matters The relationship between crime and police funding is a central topic in public policy, especially in times of economic uncertainty and shifting public perceptions of law enforcement. During the years 2020 to 2022, counties across New Jersey faced challenges brought on by the COVID-19 pandemic, including economic disruptions, rising unemployment, and changes in public safety needs.

This period presents a unique opportunity to study how counties adjusted their police budgets in response to crime trends and socioeconomic pressures. By investigating this relationship, we can better understand how local governments allocate resources in times of crisis and whether their responses align with actual public safety needs.

**Research Question**

How do crime rates in New Jersey counties influence police budget allocation?

**Hypothesis:** I hypothesize that counties with higher crime rates allocate a larger proportion of their budgets to the police, as they require more resources to address increased demand for law enforcement services.

**Key Variables**

Crime Data: Total number of reported crimes per county per year (2020–2022), indicating crime levels.

Police Budget: Total annual police department allocation per county.

County: Name of the 21 New Jersey counties included in the analysis.

Year: The time dimension, capturing data from 2020, 2021, and 2022.

Population: Estimated population per county, to provide context for per capita analysis.

Unemployment Rate: Annual unemployment rate per county, used to explore potential socioeconomic influences on crime and budget decisions.

**Data Sources**

Crime Data: New Jersey Department of Law and Public Safety.

Police Budget Data: Annual reports from the New Jersey Division of Local Government Services.

Socioeconomic Data: Population and unemployment rates from the New Jersey Department of Labor and Workforce Development.

Data Justification The integration of crime statistics, police funding data, and socioeconomic indicators enables a comprehensive analysis of how counties respond to public safety needs. Population and unemployment data provide necessary context to interpret variations in crime and budgeting patterns, while the three-year timeframe allows for the identification of temporal trends and policy shifts in response to evolving local conditions.

#**Scatter plot - Average Crime vs Average Budgets 2020 - 2022 - New Jersey**

```python
import matplotlib.pyplot as plt

# Function to clean data and convert values to float
def clean_df(df, remove_dollar=False):
    # Remove the "Total New Jersey" row
    df = df[df['County'] != 'Total New Jersey']

    # Remove commas and, if necessary, the dollar sign
    df[years] = df[years].replace({',': '', '\$': ''}, regex=True).astype(float)

    return df

# Years to consider
years = ['2020', '2021', '2022']

# Load and clean the crime and budget data (you already uploaded the URLs)
crime_df = clean_df(pd.read_csv(url_crime_data, header=1))
budget_df = clean_df(pd.read_csv(url_budget_data, header=1), remove_dollar=True)

# Calculate the averages for crime and budget
crime_df['Average_Crime'] = crime_df[years].mean(axis=1)
budget_df['Average_Budget'] = budget_df[years].mean(axis=1)

# Merge the datasets by county name
merged_df = pd.merge(crime_df[['County', 'Average_Crime']], budget_df[['County', 'Average_Budget']], on='County')

# Create the scatter plot
plt.figure(figsize=(14, 8))
plt.scatter(merged_df['Average_Crime'], merged_df['Average_Budget'], color='teal', edgecolor='black', s=100)

# Add county labels
for _, row in merged_df.iterrows():
    plt.text(row['Average_Crime'], row['Average_Budget'], row['County'],
             fontsize=8, ha='center', va='center', color='white',
             bbox=dict(facecolor='black', alpha=0.6, boxstyle='round,pad=0.2'))

# Axis labels and title
plt.title('Average Crime vs. Average Police Budget (2020–2022) - New Jersey Counties', fontsize=16)
plt.xlabel('Average Crime Count (2020–2022)', fontsize=12)
plt.ylabel('Average Police Budget (USD)', fontsize=12)

# Format the Y-axis in millions
max_budget = merged_df['Average_Budget'].max()
y_ticks = range(0, int(max_budget + 25_000_000), 50_000_000)
plt.yticks(y_ticks, [f'${x/1e6:.1f}M' for x in y_ticks])

# Show grid and adjust layout
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Interpretation:** This scatter plot displays the relationship between the average number of crimes and the average police budget across New Jersey counties from 2020 to 2022. Each point represents a county, labeled for clarity. Counties like Bergen and Essex have both high crime rates and high police budgets, reflecting larger urban populations. In contrast, counties such as Hunterdon and Sussex show lower crime numbers and receive significantly smaller budgets. There appears to be a general positive correlation—higher crime tends to be associated with higher police funding. However, some counties, like Passaic, have a relatively high budget despite moderate crime levels, suggesting other influencing factors. Smaller counties like Salem show low crime and budget levels, consistent with their size. The clustering of most counties in the lower-left quadrant suggests many regions have moderate to low crime and budget figures. Overall, the chart provides a helpful overview of how public safety funding aligns with crime trends in New Jersey.



#**Interactive Scatterplot of Average Crime vs Average Population 2020 - 2022**

```python
import plotly.express as px

# Function to clean and process data
def clean_data(df, years, remove_comma=True):
    df = df[df['County'] != 'Total New Jersey']  # Remove 'Total New Jersey'
    for year in years:
        df.loc[:, year] = df[year].str.replace(',', '', regex=True).astype(float)  # Clean data using .loc
    df['Average'] = df[years].mean(axis=1)  # Calculate average
    df['County'] = df['County'].str.strip().str.title().str.lower()  # Clean county names
    return df

# Define years to analyze
years = ['2020', '2021', '2022']

# Load and clean crime data
crime_data = pd.read_csv(url_crime_data, header=1)
crime_data = clean_data(crime_data, years)

# Load and clean population data
population_data = pd.read_csv(url_population_data)
population_data.columns = ['County', '2020', '2021', '2022']  # Renaming columns for consistency
population_data = clean_data(population_data, years)

# Merge data on 'County'
merged_data = pd.merge(crime_data[['County', 'Average']], population_data[['County', 'Average']], on='County', suffixes=('_Crime', '_Population'))

# Create the interactive scatter plot
fig = px.scatter(
    merged_data,
    x='Average_Crime',
    y='Average_Population',
    color='County',
    hover_data=['County', 'Average_Crime', 'Average_Population'],
    labels={
        'Average_Crime': 'Average Crime (2020–2022)',
        'Average_Population': 'Average Population (2020–2022)'
    },
    title='Average Crime vs. Average Population (2020–2022) by County'
)

# Customize plot for clarity
fig.update_traces(marker=dict(size=12, opacity=0.8, line=dict(width=1, color='black')))
fig.update_layout(showlegend=False)  # Hide legend to avoid clutter

# Show plot
fig.show()
```

**Interpretation:** This interactive scatter plot compares the average number of reported crimes to the average population in New Jersey counties from 2020 to 2022. Each point represents one of the 21 counties, ensuring full data coverage. Counties like Essex, Bergen, and Middlesex are located toward the top-right, showing both large populations and high crime counts. Less populated counties such as Salem, Hunterdon, and Sussex fall toward the bottom-left with lower values for both metrics. A visible positive correlation suggests that more populated areas tend to report more crimes overall. However, differences between counties with similar populations point to local factors that may influence crime rates. The use of hover labels allows for quick identification of each county’s stats. The plot is useful for identifying outliers, such as counties with higher crime levels relative to their population size. Overall, it provides a clear visual snapshot of how crime scales with population across the state.

```python
import plotly.graph_objects as go

# Load data
crime_data = pd.read_csv(url_crime_data, header=1)
budget_data = pd.read_csv(url_budget_data, header=1)
population_data = pd.read_csv(url_population_data)

# Clean 'County' column and headers
for df in [crime_data, budget_data, population_data]:
    df.columns = df.columns.str.strip()
    df['County'] = df['County'].str.strip()

# Fix column names in population_data if needed
population_data.columns = ['County', '2020', '2021', '2022']

# Convert budget to float (remove commas)
for year in ['2020', '2021', '2022']:
    budget_data[year] = budget_data[year].replace(',', '', regex=True).astype(float)
    population_data[year] = population_data[year].replace(',', '', regex=True).astype(float)

# Compute totals and averages
budget_data['Total Budget'] = budget_data[['2020', '2021', '2022']].sum(axis=1)
population_data['Population'] = population_data[['2020', '2021', '2022']].mean(axis=1)

# Merge datasets
merged = pd.merge(budget_data[['County', 'Total Budget']],
                  population_data[['County', 'Population']],
                  on='County')

# Budget per capita
merged['Budget per Capita'] = merged['Total Budget'] / merged['Population']
merged = merged.sort_values(by='Budget per Capita', ascending=False)

# Plot
fig = go.Figure(go.Bar(
    x=merged['County'],
    y=merged['Budget per Capita'],
    text=merged['Budget per Capita'].round(2),
    textposition='outside',
    marker_color='blue'
))

fig.update_layout(
    title='Police Budget per Capita by County (2020–2022)',
    xaxis_title='County',
    yaxis_title='Budget per Capita (USD)',
    xaxis_tickangle=45,
    showlegend=False,
    height=600,
    width=1000
)

fig.show()
```

#**Map of 5 top of the most dangerous cities in New Jersey**

```python
import geopandas as gpd
# Load and clean crime data
crime_data = pd.read_csv(url_crime_data, header=1)
crime_data = crime_data[crime_data['County'] != 'Total New Jersey']
for year in ['2020', '2021', '2022']:
    crime_data[year] = crime_data[year].str.replace(',', '').astype(float)

# Calculate average crime rate for 2020-2022 and get top 5 counties
crime_data['Average_2020_2022'] = crime_data[['2020', '2021', '2022']].mean(axis=1)
top5 = crime_data.nlargest(5, 'Average_2020_2022').reset_index(drop=True)

# Assign rankings and colors
top5['Rank'] = range(1, 6)
top5['Color'] = ['red', 'orange', 'yellow', 'purple', 'pink']

# Load US counties shapefile and filter for New Jersey
counties = gpd.read_file("https://www2.census.gov/geo/tiger/GENZ2022/shp/cb_2022_us_county_20m.zip")
nj_counties = counties[counties['STATEFP'] == '34']

# Clean county names for merge
nj_counties['NAME_clean'] = nj_counties['NAME'].str.strip()
top5['County_clean'] = top5['County'].str.strip()

# Merge crime data with shapefile geometry
top5_map = nj_counties.merge(top5, left_on='NAME_clean', right_on='County_clean')

# Plot the map
fig, ax = plt.subplots(figsize=(10, 12))
nj_counties.boundary.plot(ax=ax, color='black')
top5_map.plot(ax=ax, color=top5_map['Color'], edgecolor='black')

# Annotate each county with its rank and name
for _, row in top5_map.iterrows():
    centroid = row['geometry'].centroid
    ax.annotate(f"{int(row['Rank'])} - {row['County']}", xy=(centroid.x, centroid.y),
                ha='center', fontsize=9, color='black', weight='bold')

# Final map settings
ax.set_title('Top 5 Most Dangerous Counties in New Jersey (2020–2022 Avg)', fontsize=14)
ax.axis('off')
plt.show()
```

#**Conclusions and Limitations**

**Conclusions**

After conducting the analysis of crime data in New Jersey counties from 2020 to 2022, several insights emerge that tie back to the hypothesis. The scatter plot and other visualizations reveal that while there is a noticeable variation in crime rates across counties, the relationship between crime rates and factors like police budgets and unemployment is not as straightforward as expected. Although it was hypothesized that higher unemployment rates would correlate with higher crime rates, the data did not show a strong relationship between these two variables. This suggests that other factors, such as population density or local law enforcement practices, may play a more significant role in determining crime rates.

Furthermore, the analysis of crime rates across the top 5 counties with the highest total crime rates highlights that the years 2020 to 2022 saw significant variation in crime reporting. Counties like Passaic, Camden, and Essex reported some of the highest crime totals, which supports the idea that urbanized areas with larger populations tend to have higher crime rates. However, when comparing these figures with police budgets, it became clear that having a larger budget does not necessarily correlate with a reduction in crime, as some counties with high crime rates also had relatively large police budgets, challenging the assumption that more funding directly leads to lower crime.

However, there are limitations to this analysis. First, the crime data used in this study does not account for variations in crime reporting methods or law enforcement strategies, which may differ between counties. Additionally, the dataset does not include specific factors such as socioeconomic conditions or educational levels, which could influence crime rates. Lastly, the analysis did not take into account the severity or types of crime, which can vary widely and may not be equally influenced by police budgets or unemployment. Future research could focus on exploring these factors and incorporating a wider range of variables to provide a more nuanced understanding of crime trends and the effectiveness of police funding.

**Limitations and Future Research**

While the analysis provided valuable insights into crime trends across New Jersey counties, several limitations should be considered. The dataset used in this study is based on aggregate crime numbers, which may mask important nuances such as the type of crimes committed or the distribution of crime across neighborhoods. Additionally, the absence of data on law enforcement effectiveness or community-level interventions limits the ability to draw definitive conclusions about the relationship between police budgets and crime. Future research could benefit from a more granular approach that includes not only total crime rates but also information on specific crime categories, demographic factors, and community outreach programs. Furthermore, the study could be expanded by integrating longitudinal data to better understand how crime trends evolve over time and whether specific interventions have had a measurable impact.
