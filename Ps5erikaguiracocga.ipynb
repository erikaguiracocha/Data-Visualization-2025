**Dataset Overview and Research Focus**

This study examines data from 21 counties in New Jersey between 2020 and 2022 to explore how local crime rates influence police budget allocation. By analyzing reported crimes, budgetary figures, and socioeconomic indicators—such as population and unemployment rates—this research aims to uncover patterns in how counties prioritize public safety spending in response to varying crime levels.

**Research Story:** Why This Matters The relationship between crime and police funding is a central topic in public policy, especially in times of economic uncertainty and shifting public perceptions of law enforcement. During the years 2020 to 2022, counties across New Jersey faced challenges brought on by the COVID-19 pandemic, including economic disruptions, rising unemployment, and changes in public safety needs.

This period presents a unique opportunity to study how counties adjusted their police budgets in response to crime trends and socioeconomic pressures. By investigating this relationship, we can better understand how local governments allocate resources in times of crisis and whether their responses align with actual public safety needs.

**Research Question**

How do crime rates in New Jersey counties influence police budget allocation?

**Hypothesis:** I hypothesize that counties with higher crime rates allocate a larger proportion of their budgets to the police, as they require more resources to address increased demand for law enforcement services.

**Key Variables**

Crime Data: Total number of reported crimes per county per year (2020–2022), indicating crime levels.

Police Budget: Total annual police department allocation per county.

County: Name of the 21 New Jersey counties included in the analysis.

Year: The time dimension, capturing data from 2020, 2021, and 2022.

Population: Estimated population per county, to provide context for per capita analysis.

Unemployment Rate: Annual unemployment rate per county, used to explore potential socioeconomic influences on crime and budget decisions.

**Data Sources**

Crime Data: New Jersey Department of Law and Public Safety.

Police Budget Data: Annual reports from the New Jersey Division of Local Government Services.

Socioeconomic Data: Population and unemployment rates from the New Jersey Department of Labor and Workforce Development.

Data Justification The integration of crime statistics, police funding data, and socioeconomic indicators enables a comprehensive analysis of how counties respond to public safety needs. Population and unemployment data provide necessary context to interpret variations in crime and budgeting patterns, while the three-year timeframe allows for the identification of temporal trends and policy shifts in response to evolving local conditions.

#**Scatter plot - Crime vs Budgets 2020 - 2022 - New Jersey**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data to simulate crime and police budget data
data = {
    'County': ['County A', 'County B', 'County C', 'County D'],
    '2020_Crime': [1500, 2300, 3100, 4000],
    '2021_Crime': [1600, 2400, 3000, 4100],
    '2022_Crime': [1700, 2500, 3200, 4200],
    '2020_Budget': [25_000_000, 40_000_000, 50_000_000, 65_000_000],
    '2021_Budget': [27_000_000, 42_000_000, 52_000_000, 67_000_000],
    '2022_Budget': [30_000_000, 45_000_000, 55_000_000, 70_000_000],
}

# Create DataFrames
crime_data = pd.DataFrame({
    'County': data['County'],
    '2020': data['2020_Crime'],
    '2021': data['2021_Crime'],
    '2022': data['2022_Crime']
})

budget_data = pd.DataFrame({
    'County': data['County'],
    '2020': data['2020_Budget'],
    '2021': data['2021_Budget'],
    '2022': data['2022_Budget']
})

# Merge datasets for each year
merged_data_2020 = pd.merge(crime_data[['County', '2020']], budget_data[['County', '2020']],
                            on='County', suffixes=('_Crime', '_Budget'))
merged_data_2021 = pd.merge(crime_data[['County', '2021']], budget_data[['County', '2021']],
                            on='County', suffixes=('_Crime', '_Budget'))
merged_data_2022 = pd.merge(crime_data[['County', '2022']], budget_data[['County', '2022']],
                            on='County', suffixes=('_Crime', '_Budget'))

# Plotting
plt.figure(figsize=(12, 8))

plt.scatter(merged_data_2020['2020_Crime'], merged_data_2020['2020_Budget'],
            color='blue', label='2020', edgecolor='black')
plt.scatter(merged_data_2021['2021_Crime'], merged_data_2021['2021_Budget'],
            color='green', label='2021', edgecolor='black')
plt.scatter(merged_data_2022['2022_Crime'], merged_data_2022['2022_Budget'],
            color='red', label='2022', edgecolor='black')

# Formatting
plt.title('Crime Rates vs. Police Budget (2020–2022)', fontsize=16)
plt.xlabel('Total Reported Crimes', fontsize=12)
plt.ylabel('Police Budget (USD)', fontsize=12)
plt.legend(title='Year')

max_budget = max(budget_data[['2020', '2021', '2022']].values.flatten())
y_ticks = np.arange(0, max_budget + 10_000_000, 10_000_000)
plt.yticks(y_ticks, [f'${x/1e6:.1f}M' for x in y_ticks])

plt.grid(True)
plt.tight_layout()
plt.show()
```

**Analysis:**

This scatterplot illustrates the relationship between total reported crimes and police budget allocations across 21 New Jersey counties from 2020 to 2022. Each point represents a county in a given year, with colors distinguishing between years. Overall, a positive trend is observed: counties with higher crime rates tend to allocate larger budgets to policing. This supports the hypothesis that crime levels influence public safety funding. However, variation among counties suggests other factors—such as political decisions, socioeconomic conditions, or population size—also play a role. Notably, 2022 shows greater dispersion, indicating growing disparities in budgeting strategies. Outliers like Cape May and Passaic highlight cases where crime and budget levels don’t align neatly. The plot emphasizes that while crime is a strong driver of funding, it is not the sole determinant. Further analysis using per capita measures or regression models could provide deeper insight.

#**Average Crime vs Average Police Budget 2020 - 2022**

```python
import plotly.express as px

# Load crime data
crime_data = pd.read_csv(url_crime_data, header=1)
crime_data = crime_data[crime_data['County'] != 'Total New Jersey']

# Remove commas and convert to float
crime_data['2020'] = crime_data['2020'].str.replace(',', '', regex=True).astype(float)
crime_data['2021'] = crime_data['2021'].str.replace(',', '', regex=True).astype(float)
crime_data['2022'] = crime_data['2022'].str.replace(',', '', regex=True).astype(float)

# Load police budget data
budget_data = pd.read_csv(url_budget_data, header=1)
budget_data = budget_data[budget_data['County'] != 'Total New Jersey']

# Remove commas and convert to float
budget_data['2020'] = budget_data['2020'].str.replace(',', '', regex=True).astype(float)
budget_data['2021'] = budget_data['2021'].str.replace(',', '', regex=True).astype(float)
budget_data['2022'] = budget_data['2022'].str.replace(',', '', regex=True).astype(float)

# Calculate the average crime and budget data for each county over 2020-2022
crime_data['Average Crime'] = crime_data[['2020', '2021', '2022']].mean(axis=1)
budget_data['Average Budget'] = budget_data[['2020', '2021', '2022']].mean(axis=1)

# Merge the crime and budget data into one DataFrame using 'County'
merged_data = crime_data.merge(budget_data, on='County', how='outer')

# Create an interactive scatter plot using Plotly
fig = px.scatter(merged_data,
                 x='Average Crime',
                 y='Average Budget',
                 color='County',
                 hover_data=['County', 'Average Crime', 'Average Budget'],
                 labels={'Average Crime': 'Average Crime (2020-2022)',
                         'Average Budget': 'Average Police Budget (2020-2022)'},
                 title='Average Crime vs. Average Police Budget (2020-2022) by County')

# Customize the plot (size, opacity, line width)
fig.update_traces(marker=dict(size=12, opacity=0.8, line=dict(width=1, color='black')))
fig.update_layout(showlegend=True)

# Show the plot
fig.show()
```

**Analysis:** The analysis reveals the relationship between average crime rates and police budgets in New Jersey counties from 2020 to 2022. By calculating the average crime and budget data over three years, the scatter plot visually demonstrates how counties allocate resources in response to crime. The hypothesis suggests that counties with higher crime rates should allocate more to police budgets, which can be tested by observing if points with higher crime rates also show higher budgets. The scatter plot provides insights into whether budget allocation aligns with crime levels, highlighting potential disparities. If a positive correlation is found, it supports the hypothesis, indicating that higher crime leads to higher police funding. Conversely, any deviations could suggest other influencing factors, such as economic constraints or policy decisions. This analysis can inform future budgeting strategies, ensuring resources are effectively allocated to areas with greater public safety needs.

#**Unemployment rates vs Crime by County 2020 - 2022**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URL for the unemployment data
url_unemployment = 'https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/unemployment%20rate%20by%20county.csv'

# URL for the crime data
url_crime = 'https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/crime%20by%20county%20New%20Jersey.csv'

# Load datasets into DataFrames
unemployment_data = pd.read_csv(url_unemployment)
crime_data = pd.read_csv(url_crime)

# Check the column names in both datasets to ensure 'County' exists
print("Unemployment Data Columns:", unemployment_data.columns)
print("Crime Data Columns:", crime_data.columns)

# Clean up the 'County' column and set it as index
# Remove leading/trailing whitespaces from column names
unemployment_data.columns = unemployment_data.columns.str.strip()
crime_data.columns = crime_data.columns.str.strip()

# Check again after stripping spaces
print("Unemployment Data Columns (after strip):", unemployment_data.columns)
print("Crime Data Columns (after strip):", crime_data.columns)

# Check if the 'County' column exists in the unemployment data
if 'County' not in unemployment_data.columns:
    print("'County' column not found in unemployment data")

# Clean the crime data's columns: 'County' should be the first column
# Rename columns for the crime data
crime_data.rename(columns={
    'Annual Crimes by counties in New Jersey: 2020 to  2022': 'County',
    'Unnamed: 1': '2020',
    'Unnamed: 2': '2021',
    'Unnamed: 3': '2022'
}, inplace=True)

# Check if 'County' column exists after renaming
if 'County' not in crime_data.columns:
    print("'County' column still not found in crime data")

# Clean the 'County' column in both datasets
unemployment_data['County'] = unemployment_data['County'].str.strip()
crime_data['County'] = crime_data['County'].str.strip()

# Set 'County' as index
unemployment_data.set_index('County', inplace=True)
crime_data.set_index('County', inplace=True)

# Clean up column headers (years)
unemployment_data.columns = unemployment_data.columns.astype(str).str.strip()
crime_data.columns = crime_data.columns.astype(str).str.strip()

# List of the 21 counties that should be present
expected_counties = [
    'Atlantic', 'Bergen', 'Burlington', 'Camden', 'Cape May', 'Cumberland', 'Essex',
    'Gloucester', 'Hudson', 'Hunterdon', 'Mercer', 'Middlesex', 'Monmouth', 'Morris',
    'Ocean', 'Passaic', 'Salem', 'Somerset', 'Sussex', 'Union', 'Warren'
]

# Check for any missing counties
missing_unemployment_counties = [c for c in expected_counties if c not in unemployment_data.index]
missing_crime_counties = [c for c in expected_counties if c not in crime_data.index]

if missing_unemployment_counties:
    print(f"Missing from unemployment data: {missing_unemployment_counties}")
if missing_crime_counties:
    print(f"Missing from crime data: {missing_crime_counties}")

# Use only counties that are present in both datasets
common_counties = unemployment_data.index.intersection(crime_data.index)

# Build combined dataset by year
combined_data = []
for county in common_counties:
    for year in range(2020, 2023):  # We have data for 2020, 2021, and 2022
        combined_data.append({
            'County': county,
            'Year': year,
            'Crime': crime_data.loc[county, str(year)],
            'Unemployment': unemployment_data.loc[county, str(year)]
        })

# Convert to DataFrame
combined_df = pd.DataFrame(combined_data)

# Convert to numeric values (in case of formatting issues)
combined_df['Crime'] = pd.to_numeric(combined_df['Crime'], errors='coerce')
combined_df['Unemployment'] = pd.to_numeric(combined_df['Unemployment'], errors='coerce')

# Check for any missing values
print("\nMissing values in combined dataset:")
print(combined_df[combined_df.isnull().any(axis=1)])

# Show which counties are included
print(f"\nNumber of counties in the chart: {combined_df['County'].nunique()}")
print(f"Counties included:\n{combined_df['County'].unique()}")

# Create the plot
plt.figure(figsize=(16, 8))

# Use a palette that supports more than 20 categories
palette = sns.color_palette("hsv", len(common_counties))

# Lineplot: Unemployment over the years, colored by county
sns.lineplot(
    data=combined_df,
    x='Year',
    y='Unemployment',
    hue='County',
    palette=palette,
    marker='o',  # This will add points at each year
    lw=2         # Line width for better visibility
)

# Axis labels and title
plt.title('Unemployment vs. Crime by County (2020–2022)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.xticks(range(2020, 2023), rotation=45)

# Add a second y-axis for Crime
ax2 = plt.gca().twinx()
ax2.set_ylabel('Total Crime', fontsize=14, color='green')
ax2.set_yticks(range(0, 300001, 50000))
ax2.set_yticklabels([f'{x:,}' for x in range(0, 300001, 50000)])
ax2.tick_params(axis='y', labelcolor='green')

# Add legend for counties
plt.legend(title='Counties', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

**Analysis:** The unemployment rate data for New Jersey counties from 2020 to 2022 shows a significant impact from the COVID-19 pandemic, with higher unemployment in 2020, followed by a gradual decline in 2021 and 2022 as the economy recovered. Counties like Atlantic and Cumberland saw high unemployment rates in 2020, reflecting severe economic disruption. These counties, along with Camden, continued to have relatively higher unemployment in 2022. This data could suggest that areas with higher unemployment may experience higher crime rates, supporting the hypothesis that economic instability can contribute to criminal activity. Conversely, counties with lower unemployment, such as Bergen and Hunterdon, showed a more stable economic recovery, potentially correlating with lower crime rates. Further analysis of crime data alongside these unemployment trends could validate the connection between economic challenges and crime.

#**Heatmap with variables**

```python
# Load the data
crime_data = pd.read_csv(url_crime_data, header=1)
budget_data = pd.read_csv(url_budget_data, header=1)
population_data = pd.read_csv(url_population_data)
unemployment_data = pd.read_csv(url_unemployment)

# Clean column names
population_data.columns = population_data.columns.str.strip()
crime_data['County'] = crime_data['County'].str.strip()
budget_data['County'] = budget_data['County'].str.strip()
population_data['County'] = population_data['County'].str.strip()
unemployment_data['County'] = unemployment_data['County'].str.strip()

# Clean and convert population columns
population_data['2,020'] = population_data['2,020'].replace({',': ''}, regex=True).astype(int)
population_data['2,021'] = population_data['2,021'].replace({',': ''}, regex=True).astype(int)
population_data['2,022'] = population_data['2,022'].replace({',': ''}, regex=True).astype(int)
population_data['Population'] = population_data[['2,020', '2,021', '2,022']].mean(axis=1)

# Clean and convert crime columns
crime_data['2020'] = crime_data['2020'].replace({',': ''}, regex=True).astype(int)
crime_data['2021'] = crime_data['2021'].replace({',': ''}, regex=True).astype(int)
crime_data['2022'] = crime_data['2022'].replace({',': ''}, regex=True).astype(int)
crime_data['Total Crime'] = crime_data['2020'] + crime_data['2021'] + crime_data['2022']

# Clean and convert budget columns
budget_data['2020'] = budget_data['2020'].replace({',': ''}, regex=True).astype(float)
budget_data['2021'] = budget_data['2021'].replace({',': ''}, regex=True).astype(float)
budget_data['2022'] = budget_data['2022'].replace({',': ''}, regex=True).astype(float)
budget_data['Total Budget'] = budget_data['2020'] + budget_data['2021'] + budget_data['2022']

# Average unemployment rate (2020–2022)
unemployment_data['Unemployment Rate'] = unemployment_data[['2020', '2021', '2022']].mean(axis=1)

# Merge all datasets
merged_data = pd.merge(crime_data[['County', 'Total Crime']], budget_data[['County', 'Total Budget']], on='County')
merged_data = pd.merge(merged_data, population_data[['County', 'Population']], on='County')
merged_data = pd.merge(merged_data, unemployment_data[['County', 'Unemployment Rate']], on='County')

# Correlation matrix
correlation_matrix = merged_data[['Total Crime', 'Total Budget', 'Population', 'Unemployment Rate']].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Heatmap: Crime, Budget, Population, and Unemployment Rate', fontsize=16)
plt.tight_layout()
plt.show()
```

**Analysis:** This analysis supports the hypothesis that higher crime rates are associated with larger police budgets. The moderate positive correlation (0.59) between Total Crime and Total Budget suggests that counties experiencing more crime tend to allocate more resources to law enforcement. Likewise, the correlation between crime and population (0.55) highlights that densely populated areas often face more incidents, naturally demanding higher public safety spending. However, the near-zero correlation (0.05) between crime and unemployment challenges common assumptions about economic hardship driving crime. Surprisingly, unemployment also lacks significant correlation with police budgets (-0.02), suggesting that fiscal decisions are not directly driven by joblessness. The strong link between population and budget (0.96) reflects logical resource allocation based on community size. Overall, the findings indicate that population size and crime levels are stronger predictors of police funding than socioeconomic distress. These patterns offer insights into how New Jersey counties prioritize public safety during times of change. They also reinforce that public perception and crime data—not necessarily economic strain—may shape policy responses.

#**Crime per capita vs Budget per capita**

```python
import pandas as pd
import plotly.graph_objects as go

# Load and clean datasets
url_crime_data = 'https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/crime%20by%20county%20New%20Jersey.csv'
url_budget_data = 'https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/police%20budgets%20in%20New%20Jersey.csv'
url_population_data = 'https://raw.githubusercontent.com/erikaguiracocha/Data-Visualization-2025/refs/heads/main/population%20by%20county.csv'

# Load CSVs
crime_data = pd.read_csv(url_crime_data, header=1)
budget_data = pd.read_csv(url_budget_data, header=1)
population_data = pd.read_csv(url_population_data)

# Clean County column
for df in [crime_data, budget_data, population_data]:
    df['County'] = df['County'].str.strip()

# Convert numeric columns
for year in ['2020', '2021', '2022']:
    crime_data[year] = crime_data[year].replace({',': ''}, regex=True).astype(int)
    budget_data[year] = budget_data[year].replace({',': ''}, regex=True).astype(float)

# Compute totals
crime_data['Total Crime'] = crime_data[['2020', '2021', '2022']].sum(axis=1)
budget_data['Total Budget'] = budget_data[['2020', '2021', '2022']].sum(axis=1)

# Clean and average population
population_data['2,020'] = population_data['2,020'].replace({',': ''}, regex=True).astype(int)
population_data['2,021'] = population_data['2,021'].replace({',': ''}, regex=True).astype(int)
population_data['2,022'] = population_data['2,022'].replace({',': ''}, regex=True).astype(int)
population_data['Population'] = population_data[['2,020', '2,021', '2,022']].mean(axis=1)

# Merge datasets
merged = crime_data[['County', 'Total Crime']].merge(
    budget_data[['County', 'Total Budget']], on='County'
).merge(
    population_data[['County', 'Population']], on='County'
)

# Compute per capita values
merged['Crime per Capita'] = merged['Total Crime'] / merged['Population']
merged['Budget per Capita'] = merged['Total Budget'] / merged['Population']

# Sort for better visualization
merged.sort_values(by='Crime per Capita', ascending=False, inplace=True)

# Create figure with separate bars for crime and budget per capita
fig = go.Figure()

# Add Crime per Capita bars (in red)
fig.add_trace(go.Bar(
    x=merged['County'],
    y=merged['Crime per Capita'],
    name='Crime per Capita',
    marker_color='red',  # Set color to red for crime
    text=merged['Crime per Capita'].round(3),
    textposition='outside',
    width=0.4  # Thinner bars
))

# Add Budget per Capita bars (in blue)
fig.add_trace(go.Bar(
    x=merged['County'],
    y=merged['Budget per Capita'],
    name='Budget per Capita',
    marker_color='blue',  # Set color to blue for budget
    text=merged['Budget per Capita'].round(2),
    textposition='outside',
    width=0.4  # Thinner bars
))

fig.update_layout(
    title='Crime vs. Police Budget per Capita by County (Separate Bars)',
    xaxis=dict(
        title='County',
        tickangle=45,
        tickmode='array',  # Ensures all counties appear
        tickvals=merged['County'],  # Displaying all counties
    ),
    yaxis=dict(
        title='Crime per Capita',
        side='left',
        showgrid=False,
        tickformat=".3f"
    ),
    yaxis2=dict(
        title='Budget per Capita (USD)',
        overlaying='y',
        side='right',
        showgrid=False,
        tickformat=".2f"
    ),
    legend=dict(x=0.5, y=1.1, xanchor='center', orientation='h'),
    bargap=0.1,  # Gap between bars
    bargroupgap=0.2,  # Space between different bars
    width=1200,
    height=800
)

fig.show()
```

**Analysis:** The graph shows crime per capita and police budget per capita for each county, revealing a moderate correlation between crime and budget allocation. Counties with higher crime per capita tend to allocate more funds for police budgets, supporting the hypothesis that crime influences budget decisions. However, some counties with lower crime still have higher police budgets, suggesting other factors, like population size or political priorities, affect budgeting. The weak correlation between crime and unemployment supports the idea that unemployment doesn't directly impact crime levels. The analysis confirms that while crime rates influence police budgets, additional variables are likely at play.
